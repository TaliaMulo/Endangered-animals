```python
def replace_missing_values(df, col_to_def_val_dict):
    dfReplace=df.copy()
    list_num=df.select_dtypes(include="number").columns
    for x in df.columns:
        if x in list_num:
            dfReplace[x].fillna(df[x].median(), inplace=True)
    return dfReplace
```

# Machine Learning

```python
def get_classifier_obj(classifier_name, params):
    if classifier_name=='KNN':
        clf=KNeighborsClassifier(n_neighbors=params['n_neighbors'])
    elif classifier_name=='decision_tree':
        clf=tree.DecisionTreeClassifier(max_depth=params['max_depth'],min_samples_split=params['min_samples_split'])
    elif classifier_name=='random_forest':
        clf=RandomForestClassifier(n_estimators=params['n_estimators'])
    elif classifier_name=='naive_bayes':
        clf=GaussianNB()
    elif classifier_name=='svm':
        clf=svm.SVC(kernel='linear')
    elif classifier_name=='logistic_regression':
        clf=LogisticRegression(solver='liblinear')
            
    return clf
```

```python
def calc_score(eval_metric, y_test, y_predicted):
    if eval_metric=='accuracy':
        evaluation_val=accuracy_score(y_test,y_predicted)
    elif eval_metric=='precision':
        evaluation_val=precision_score(y_test,y_predicted,average = 'macro')
    elif eval_metric=='recall':
        evaluation_val=recall_score(y_test,y_predicted,average = 'macro')
    elif eval_metric=='f1':
        evaluation_val=f1_score(y_test,y_predicted,average = 'macro')
        
    return evaluation_val
```

## Train Test and Split the Dataset


```python
def split_to_train_and_test(dataset, label_column, test_ratio, rand_state):
    numeric_columns = dataset._get_numeric_data().columns
    x=dataset[numeric_columns].drop(label_column,axis=1)
    return train_test_split(x, dataset[label_column], test_size=test_ratio, random_state=rand_state)
```

```python
test_ratio, rand_state = 0.2, 42
category_col_name = 'Danger_of_extinction'
X_train, X_test, y_train, y_test = split_to_train_and_test(df2, category_col_name, test_ratio, rand_state)
```

```python
def train_model(model,X_train, y_train):
        return model.fit(X_train, y_train)
```

```python
def predict(trained_model, X_test):
    y_pred = trained_model.predict(X_test)
    return y_pred
```

```python
data_score = {
'Model':[] ,
'Accuracy': [],
'Precision': [],
'Recall':[],
'f1': [] }
```

## Models
### KNN 

```python
params_knn = {'n_neighbors':71}
model = get_classifier_obj('KNN', params_knn)

train = train_model(model,X_train, y_train)

y_pred = predict(train, X_test)
y_pred_train = predict(train, X_train)

a_train = calc_score("accuracy", y_train, y_pred_train)
accuracy_val = calc_score("accuracy", y_test, y_pred)
precision_val = calc_score("precision", y_test, y_pred)
recall_val = calc_score("recall", y_test, y_pred)
f1_val = calc_score("f1", y_test, y_pred)

print('Accuracy on train:',a_train)
print('Accuracy on test:',accuracy_val)
print('Precision:', precision_val)
print('Recall:',recall_val)
print('f1:', f1_val)
```

```python
data_score['Model'].append('KNN')
data_score['Accuracy'].append(accuracy_val)
data_score['Precision'].append(precision_val)
data_score['Recall'].append(recall_val)
data_score['f1'].append(f1_val)
```
