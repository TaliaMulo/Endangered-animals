# Crawling

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
import time

def get_soup(url):
    user_agent={'User-agent': 'Mozilla/5.0'}
    response=requests.get(url,headers=user_agent)
    soup=BeautifulSoup(response.content,"html.parser", from_encoding="utf-8")
    return soup

def dump_urls():
    soup1=get_soup('https://a-z-animals.com/animals/')
    names_animal=[]
    links=[]
    newsoup1 = soup1.findAll('li', attrs={'class': 'list-item'})
    for i in newsoup1:
        names_animal.append(i.find("a").get_text().strip())
        links.append(i.find('a').get('href'))
    pd.DataFrame({'animal': names_animal, 'url': links}).to_csv('animal_to_url_mapping.csv', index=False)


def build_data():
    map_df = pd.read_csv('animal_to_url_mapping.csv')
    links = map_df['url'].tolist()
    names = map_df['animal'].tolist()
    if os.path.exists('data_dump.txt'):
        with open('data_dump.txt', 'r', encoding='utf-8') as rf:
            links_that_already_dumped = [l.replace('url:', '') for l in rf.read().splitlines() if 'url:' in l]
    else:
        links_that_already_dumped = []

    with open('data_dump.txt', 'a', encoding='utf-8') as wf:
        for url, name in zip(links, names):
            if url in links_that_already_dumped:
                print(f'url {url} already dumped')
            else:
                print(url)
                wf.write(f'*\n')
                wf.write(f'url:{url}\n')
                wf.write(f'name:{name}\n')
                soup2=get_soup(url)
                newsoup2=soup2.find('div',attrs={'class': 'row'})
                for i,h in zip(newsoup2('dt'), newsoup2('dd')):
                    feature = i.get_text()
                    value = h.get_text()
                    try:
                        wf.write(f'{feature}:{value}\n')
                    except:
                        print('cant write feature')
                label = newsoup2.find('ul', attrs={'class':'list-unstyled'}).get_text()
                wf.write(f'label:{label}\n')
                time.sleep(2)

                    
```

```python
dump_urls()
```

```python
build_data()
```
